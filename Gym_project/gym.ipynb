```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('megaGymDataset.csv',index_col='Title')
df.drop(columns=['Unnamed: 0'],inplace=True)
#Display Cardio exercises with highest rating
df[df['Type']=='Cardio'].sort_values('Rating',ascending=False).head(10)
#Display Strength exercises 
```

```python
# Retrieve rows where Type is 'Strength' and BodyPart is 'Biceps'
biceps_str = df[(df['Type']=='Strength') & (df['BodyPart']=='Biceps')].sort_values('Rating', ascending=False).head(10)

sns.barplot(data=biceps_str[0:10], x='Rating', y=biceps_str.index, palette='viridis')
plt.title('Biceps Strength Exercises by Rating')
top_10_biceps = [i for i in biceps_str.index]
top_10_biceps_rating = [i for i in biceps_str['Rating']]
```

```python
import json

with open('top_10_biceps.json', 'w') as f:
    json.dump(top_10_biceps, f)
```
